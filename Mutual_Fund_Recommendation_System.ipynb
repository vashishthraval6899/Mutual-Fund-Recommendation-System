# **2. Feature Descriptions**


| **Column Name**                 | **Description**                                                                                                                                                |
| ------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Funds**                       | The official name of the mutual fund scheme.                                                                                                                   |
| **Fund Manager**                | The person responsible for making investment decisions and managing the funds portfolio.                                                                      |
| **Category**                    | The classification of the mutual fund based on its investment strategy (e.g., Equity, Hybrid, Debt).                                                           |
| **RupeeVestRating**             | Rating assigned by RupeeVest (on a scale or as Unrated), indicating the funds relative performance and risk-adjusted returns.                               |
| **AUM (in Rs. cr)**             | Assets Under Management  the total market value of all assets held by the fund, measured in crores of rupees. It indicates fund size and investor confidence. |

| **Fund Type**                   | Indicates whether the fund is *Open Ended* (can be bought/sold anytime) or *Closed Ended* (available only during specific periods).                            |
| **Inception Date**              | The date when the fund was launched. Helps in evaluating the funds track record.                                                                              |
| **Benchmark Index**             | The market index used to compare the funds performance (e.g., NIFTY 50, CRISIL Bond Index).                                                                   |
| **NAV**                         | Net Asset Value  the per-unit price of the fund; calculated as (Total Assets - Liabilities) / Number of Units.                                                |
| **52 Week High (NAV)**          | The highest Net Asset Value of the fund during the past 52 weeks.                                                                                              |
| **52 Week Low (NAV)**           | The lowest Net Asset Value of the fund during the past 52 weeks.                                                                                               |
| **Return (%) 1 mo**             | Percentage return generated by the fund over the last 1 month.                                                                                                 |
| **Return (%) 3 mo**             | Percentage return generated by the fund over the last 3 months.                                                                                                |
| **Return (%) 6 mo**             | Percentage return generated by the fund over the last 6 months.                                                                                                |
| **Return (%) 1 yr**             | Percentage return generated by the fund over the last 1 year.                                                                                                  |
| **Return (%) 2 yrs**            | Annualized return of the fund over the last 2 years.                                                                                                           |
| **Return (%) 3 yrs**            | Annualized return of the fund over the last 3 years.                                                                                                           |
| **Return (%) 5 yrs**            | Annualized return of the fund over the last 5 years.                                                                                                           |
| **Return (%) 10 yrs**           | Annualized return of the fund over the last 10 years.                                                                                                          |
| **PE Ratio**                    | Price-to-Earnings Ratio  indicates how much investors are paying for 1 of the funds earnings. Lower values may indicate undervaluation.                     |
| **PB Ratio**                    | Price-to-Book Ratio  compares the market value to the book value of the funds holdings; used to assess valuation levels.                                     |
| **Turnover Ratio (%)**          | Measures how frequently securities within the fund are traded in a year; higher ratios imply more active management.                                           |
| **No. of Stocks**               | The number of individual stock holdings in the funds portfolio.                                                                                               |
| **Avg. Market Cap (in Rs. cr)** | The average market capitalization of the companies in the fund; shows whether it focuses on large-cap, mid-cap, or small-cap stocks.                           |
| **Large Cap (%)**               | Percentage of the portfolio invested in large-cap companies.                                                                                                   |
| **Mid Cap (%)**                 | Percentage of the portfolio invested in mid-cap companies.                                                                                                     |
| **Small Cap (%)**               | Percentage of the portfolio invested in small-cap companies.                                                                                                   |
| **Highest Sector**              | The sector with the largest allocation within the fund (e.g., Financials, Technology).                                                                         |

# **11. Hypothesis Testing**

# **Test 1: Return vs Category**
# H: Mean 3Y return is same across categories
# H: At least one category differs

```python
# Step 1: Group data by category
groups = [group['Return (%)3 yrs'].values for name, group in data.groupby('Category')]
```

```python
# Step 2: Perform one-way ANOVA
f_stat, p_val = stats.f_oneway(*groups)
```

```python
# Step 3: Print results
print("ANOVA Test: Return (%)3 yrs across Categories")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.6f}")
```

```python
# Step 4: Interpretation
alpha = 0.05
if p_val < alpha:
    print(" Reject H: Mean 3-year returns differ significantly among categories.")
else:
    print(" Fail to reject H: No significant difference among category means.")
```

# **12. Feature Engineering**

# **12.1 Return Momentum Features**
- These capture how returns change over time, not just the final level.

```python
# Funds with similar average return but different paths will now separate better.
data['Mom_1_6'] = data['Return (%)6 mo'] - data['Return (%)1 mo']
data['Mom_3_12'] = data['Return (%)1 yr'] - data['Return (%)3 mo']
data['Mom_6_36'] = data['Return (%)3 yrs'] - data['Return (%)6 mo']
```

# **12.2  Risk & Volatility Features**
 - These capture how risky the fund is and whether the risk is good (rewarded) or bad (downside).

```python
data['Volatility'] = data['Standard Deviation']

# Average Annual Return (AAR)
data['Avg_Return'] = data[['Return (%)1 yr','Return (%)2 yrs','Return (%)3 yrs']].mean(axis=1)

#Helps cluster high-return/low-risk vs low-return/high-risk funds.
data['Return_Risk_Ratio'] = data['Avg_Return'] / (data['Standard Deviation'] + 1e-6)

#This gives extra separation between funds with same total risk but different risk quality.
data['Downside_Risk'] = data['Standard Deviation'] * (1 - data['Sharpe'])

#Comparing Sortino to Sharpe tells us: How much of total risk is actually bad risk?
data['Up_Down_Ratio'] = data['Sortino'] / (data['Sharpe'] + 1e-6)
```

# **12.3 Risk-Adjusted Performance**
- These combine skill measures like Alpha, Sharpe, Sortino, and Beta.

```python
#High alpha with low beta is more impressive than same alpha with very high beta.
data['Alpha_Beta_Score'] = data['Alpha'] / (data['Beta'] + 1e-6)

#Summing them gives a quick overall quality score.
data['Sharpe_Sortino_Score'] = data['Sharpe'] + data['Sortino']
```

# **12.4 Size & Cost Features**

```python
#Bigger funds tend to be more stable but sometimes less aggressive.
data['AUM_Score'] = np.log(data['AUM(in Rs. cr)'] + 1)

#Clusters can then separate cheap but good vs expensive but underperforming funds.
data['Cost_Efficiency'] = 1 / (data['ExpenseRatio (%)'] + 1e-6)
```

# **12.5 Class of Mutual Fund**

```python
# Create a class feature from classification to make identification easier
data['Class']=data['classification'].str.split().str[0]
```

```python
#Lets see in each class how many funds are there
data['Class'].value_counts().plot(kind='bar')

# Add labels and title
plt.title('Class Frequency', fontsize=14)
plt.xlabel('Class Name', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the chart
plt.show()
```

# **12.6 Category-Level Features (within Class)**

```python
# Risk Score = average of normalized Beta and Std Dev
data['Risk_Score'] = (data['Beta'].rank(pct=True) + data['Standard Deviation'].rank(pct=True)) / 2

#Z-score of Avg_Return within each Class.
data['Class_Return_Z'] = data.groupby('Class')['Avg_Return'].transform(lambda x: (x - x.mean()) / (x.std() + 1e-6))

#Z-score of Risk_Score within each Class.
data['Class_Risk_Z'] = data.groupby('Class')['Risk_Score'].transform(lambda x: (x - x.mean()) / (x.std() + 1e-6))

#Z-score of Standard Deviation within Class.
data['Class_SD_Z'] = data.groupby('Class')['Standard Deviation'].transform(lambda x: (x - x.mean()) / (x.std() + 1e-6))
```

```python
# Choose best k by max silhouette
best_k_kmeans = K[np.argmax(silhouette_scores)]
print("Best k for KMeans (by silhouette):", best_k_kmeans)

# Fit final KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=50)
kmeans_labels = kmeans.fit_predict(data1)
print("KMeans silhouette:", silhouette_score(data1, kmeans_labels))
```

# **14.3 DBSCAN**

```python
#Performing Hyperparameter Tuning
from tqdm import tqdm

def tune_dbscan(X, eps_values, min_samples_values):
    results = []

    for eps in eps_values:
        for min_samples in min_samples_values:

            db = DBSCAN(eps=eps, min_samples=min_samples)
            labels = db.fit_predict(X)

            # Noise mask
            valid_mask = labels != -1
            unique_clusters = np.unique(labels[valid_mask])

            # Need at least 2 clusters for silhouette
            if len(unique_clusters) < 2:
                sil = -1
            else:
                sil = silhouette_score(X[valid_mask], labels[valid_mask])

            results.append({
                "eps": eps,
                "min_samples": min_samples,
                "clusters_found": len(unique_clusters),
                "noise_points": np.sum(labels == -1),
                "silhouette": sil
            })

    return pd.DataFrame(results)

# RUN TUNING

eps_range = np.arange(0.2, 5.0, 0.05)
min_samples_range = [3, 4, 5, 6, 7, 8]

dbscan_results = tune_dbscan(data1, eps_range, min_samples_range)

# Sort by silhouette score
best_dbscan = dbscan_results.sort_values("silhouette", ascending=False).head(10)
print(best_dbscan)
```

# **14.4 Gaussian Mixture Model (GMM)**

```python
bic_scores_gmm = []
sil_scores_gmm = []
K_gmm = range(2, 15)

for k in K_gmm:
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm.fit(data1)
    labels_gmm_k = gmm.predict(data1)

    bic_scores_gmm.append(gmm.bic(data1))
    sil_scores_gmm.append(silhouette_score(data1, labels_gmm_k))

# Plot BIC & Silhouette
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(K_gmm, bic_scores_gmm, 'bo-')
ax1.set_title("GMM: BIC Scores (lower = better)")
ax1.set_xlabel("Number of components"); ax1.set_ylabel("BIC")
ax1.grid(True)

ax2.plot(K_gmm, sil_scores_gmm, 'ro-')
ax2.set_title("GMM: Silhouette Scores")
ax2.set_xlabel("Number of components"); ax2.set_ylabel("Silhouette Score")
ax2.grid(True)

plt.tight_layout()
plt.show()

best_k_gmm = K_gmm[np.argmax(sil_scores_gmm)]
print("Best n_components for GMM (by silhouette):", best_k_gmm)

gmm = GaussianMixture(n_components=best_k_gmm, random_state=42)
gmm_labels = gmm.fit_predict(data1)
print("GMM silhouette:", silhouette_score(data1, gmm_labels))
```

```python
# Quality score to rank funds within clusters

# Build a composite quality score from engineered features
quality_components = {
    "Sharpe_Sortino_Score": 0.25,
    "Return_Risk_Ratio":    0.20,
    "Fund_Strength":        0.15,
    "Alpha_Beta_Score":     0.10,
    "Mom_3_12":             0.10,
    "Mom_6_36":             0.10,
    "Cost_Efficiency":      0.05,
    "Volatility":          -0.05,  # penalize higher volatility
}

quality_score = np.zeros(len(data))

for col, w in quality_components.items():
    quality_score += w * data[col].values

data["Quality_Score_raw"] = quality_score

# Normalize quality score to [0, 1] for interpretability
mm = MinMaxScaler()
data["Quality_Score"] = mm.fit_transform(data[["Quality_Score_raw"]])
```

```python
# get top funds in a cluster

def get_top_funds_in_cluster(df, cluster_id, top_n=20):
    """
    Return the top N funds in a given cluster based on Quality_Score.
    """
    cluster_df = df[df["cluster"] == cluster_id].copy()
    cluster_df = cluster_df.sort_values("Quality_Score", ascending=False)

    display_cols = [
        "Funds", "Class", "cluster",
        "AUM(in Rs. cr)", "ExpenseRatio (%)", "NAV",
        "Return (%)1 yr", "Return (%)2 yrs", "Return (%)3 yrs",
        "Beta", "Standard Deviation",
        "Sharpe", "Sortino",
        "Avg_Return", "Risk_Score", "Risk_Adjusted",
        "Fund_Strength", "Quality_Score"
    ]

    # Keep only columns that exist in df
    display_cols = [c for c in display_cols if c in cluster_df.columns]

    return cluster_df[display_cols].head(top_n)
```

```python
#Storing top 20 funds in each clusters
top_cluster_0 = get_top_funds_in_cluster(data, cluster_id=0, top_n=20)
top_cluster_1 = get_top_funds_in_cluster(data, cluster_id=1, top_n=20)
top_cluster_2 = get_top_funds_in_cluster(data, cluster_id=2, top_n=20)
```
